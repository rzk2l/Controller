```python
import tkinter as tk
import socket
import struct
import keyboard

kp = 0
ki = 0
kd = 0
new_throttle = 1000
killFlag = 0

def on_up_arrow(event=None):
    global new_throttle
    new_throttle += 50
    if new_throttle > 2000:
        new_throttle = 2000
    print("New throttle value:", new_throttle)
    send_throttle()

def on_down_arrow(event=None):
    global new_throttle
    new_throttle -= 50
    if new_throttle < 1000:
        new_throttle = 1000
    print("New throttle value:", new_throttle)
    send_throttle()

def on_s_key(event=None):
    global killFlag
    killFlag = 1
    print("Killed speed")
    send_throttle()
    exit()

def send_throttle():
    global new_throttle, kp, ki, kd, killFlag

    HOST = '192.168.43.89' # Replace with the IP address of your ESP32
    PORT = 80

    # Create a socket connection to the ESP32
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.connect((HOST, PORT))
        
        # Pack the throttle, kp, ki, and kd values into a struct
        #print(type(kp),type(ki),type(kd))
        packed_data = struct.pack('<HfffH', new_throttle, float(kp), float(ki), float(kd),killFlag)
        
        # Send the packed data to the ESP32
        s.sendall(packed_data)

        # Receive the new throttle value from the server
        data = s.recv(2)
        new_throttle = struct.unpack('H', data)[0]
        
        #print('New throttle value:', new_throttle)

def on_kp_changed(new_value):
    global kp
    kp = new_value
    print("New kp value:", kp)
    send_throttle()

def on_ki_changed(new_value):
    global ki
    ki = new_value
    print("New ki value:", ki)
    send_throttle()

def on_kd_changed(new_value):
    global kd
    kd = new_value
    print("New kd value:", kd)
    send_throttle()

def create_gui():
    root = tk.Tk()          #Calls a constructor, constructs a window
    root.geometry("800x600")
    root.title("Tunator600")

    titleLabel = tk.Label(root, text="Tuner")
    titleLabel.pack(padx=30,pady=20)

    # Create a label and a slider for kp
    kp_label = tk.Label(root, text="Kp :")
    kp_label.pack()
    kp_slider = tk.Scale(root, from_=0, to=3, resolution=0.1, orient=tk.HORIZONTAL,length=700,tickinterval=0.2, command=on_kp_changed)
    kp_slider.pack()
    kp_slider.set(kp)

    # Create a label and a slider for ki
    ki_label = tk.Label(root, text="ki:")
    ki_label.pack()
    ki_slider = tk.Scale(root, from_=0, to=1, resolution=0.01, orient=tk.HORIZONTAL,length=700,tickinterval=0.10, command=on_ki_changed)
    ki_slider.pack()
    ki_slider.set(ki)

    # Create a label and a slider for kd
    kd_label = tk.Label(root, text="kd:")
    kd_label.pack()
    kd_slider = tk.Scale(root, from_=0, to=70, resolution=1, orient=tk.HORIZONTAL,length=700,tickinterval=10, command=on_kd_changed)
    kd_slider.pack()
    kd_slider.set(kd)
    
    keyboard.add_hotkey('up', on_up_arrow)
    keyboard.add_hotkey('down', on_down_arrow)
    keyboard.add_hotkey('s', on_s_key)

    root.mainloop()                  #Run the app

create_gui()
# KP = 1.5
# KI = 0.03
# KD = 60
```
